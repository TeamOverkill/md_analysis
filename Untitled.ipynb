## 1. The Maxwell-Boltzmann Distrubtion: Exact vs. Simulation
A MD simulation starts from it's intial conditions, which includes it's initial configuration (the placement of particles) and it's initial velocities. The speed of the particles is related to the kinetic energy of the system
$$E_{K} = \sum_{i}^{N} \frac{m_i v_i^2}{2} \tag{1.1}$$
where  $m_i$ the mass and $v_i$ the speed of particle $i$. The average kinetic energy is realted to the temperature through the equipartition theorem
$$2 \langle E_{K} \rangle = g k_{B} T \stackrel{\mathrm{3D}}{=} 3N k_{B}T \tag{1.2}$$ 
with $k_{B}$ being the Boltzmann constant, $T$ the absolute temperature and $g$ the number of translational degrees of freedom with the last equally holding for an ideal gas in a three-dimensional space. Consequently if we know the desired temperature, $T$, for the system, we can assign velocities to the particles according to the Maxwell-Boltzmann distribution
$$P(\mathbf{v}_{i})= \sqrt{\frac{m_i}{2\pi k_{B}T}} \exp{\left( {-\frac{m_i \mathbf{v}_{i}}{k_B T}}\right)} \tag{1.3}$$
Note here the Maxwell-Boltzmann distribution is expressed in terms of vectorial velocities ($\mathbf{v}$), unlike previously where we adressed the particle's speed ($v$) (absolute value of the velocity). Realizing that eq. 1.3 is simply a [Gaussian distribution](https://en.wikipedia.org/wiki/Normal_distribution#Standard_normal_distribution) a mean equal to 0 and variance equal to
$$\sigma_{i}^2 = \frac{k_B T}{m_{i}} \tag{1.4}$$
The problem boils down to generating random numbers from a gaussian distribution, as the Maxwell-Boltzmann distribution can be written as
$$\mathbf{v}_{i}=\sqrt{\frac{k_B T}{m}} \mathcal{N}(0,1) \tag{1.5}$$
In order to generate random gaussian distributed numbers with mean 0 and variance 1 from randomly uniformed numbers between 0 and 1, we conduct a [Box‚ÄìMuller transformation](https://en.wikipedia.org/wiki/Box‚ÄìMuller_transform#Basic_form) of the randomly uniformed numbers $U_{1}$ and $U_{2}$ into standard normal distributed random numbers $Z_{1}$ and $Z_{2}$, through eq. 1.6 and 1.7
$$Z_{1} = \sqrt{-2 \ln U_{1}} \cos(2\pi U_{2}) \tag{1.6}$$
$$Z_{2} = \sqrt{-2 \ln U_{2}} \cos(2\pi U_{1}) \tag{1.7}$$
<br/><br/>
In the code below we have illustrated how this procedure can successfully generate initial Maxwell-Boltzmann distributed velocities for a nitrogen gas using the Molecular Overkill Engine compared to the analytical solution.

```python
data = np.loadtxt('velocities.txt', usecols=1, unpack=True)
data_unit = data * 1000    # nm/ps -> m/s

# Maxwell-Boltzmann distribution
m = 2*14.0067*1.660538921e-27  # Mass (atomic mass units * dalton->kg)
T = 298.15                     # Temperature (Kelvin)
kB = 1.3806503e-23             # Boltzmann constant (J/K)
def fmb(m, T, v):
    return (m/(2.*np.pi*kB*T))**1.5 * 4.*np.pi*v*v * np.exp((-m*v*v)/(2.*kB*T))

v = np.arange(0.0, 1500, 0.01)


# Initialize figure and axes
fig = plt.figure(figsize=(8.4,6))
ax1 = fig.add_subplot(111)

# Plot Data
ax1.plot(v, fmb(m, T, v), 'r--', linewidth=3)
ax1.hist(data_unit, bins=50, density=True, facecolor='green', alpha=0.70, edgecolor='purple');

# Axis modification
ax1.set_xlim(0,1500)
ax1.minorticks_on()
ax1.tick_params(axis='both',which='minor',length=5,width=2,labelsize=18)
ax1.tick_params(axis='both',which='major',length=8,width=2,labelsize=18)

# Labels and text
ax1.set_xlabel(r'Speed (m/s)', fontsize=22, labelpad=20)
ax1.set_ylabel(r'Probability', fontsize=22, labelpad=20)
ax1.set_title(r'Maxwell-Boltzmann Distribution of N$_{2}$', fontsize=22)
ax1.text(650, 0.0015, r'$f(v) = \sqrt{(\frac{m}{2 \pi k_{B} T})^{3}} 4\pi v^{2} e^{{\frac{-mv^{2}}{2k_{B}T}}}$', fontsize=20, color='b')

plt.tight_layout()
#fig.savefig('Distribution.png', hbox_inches='tight', dpi=600)
plt.show()
Markdown('''**Figure 1.1**: Histogram of the initial speed distribution of nitrogen gas (N={:.2e})
            from a molecular dynamics simulation (green) compared to the analytical 
            Maxwell-Boltzmann distribution (<span style="color:red">**--**</span>)
            with the equation written in the plot.'''.format(len(data_unit)))
```
