## 1. The Maxwell-Boltzmann Distrubtion: Exact vs. Simulation
A MD simulation starts from it's intial conditions, which includes it's initial configuration (the placement of particles) and it's initial velocities. The speed of the particles is related to the kinetic energy of the system
$$E_{K} = \sum_{i}^{N} \frac{m_i v_i^2}{2} \tag{1.1}$$
where  $m_i$ the mass and $v_i$ the speed of particle $i$. The average kinetic energy is realted to the temperature through the equipartition theorem
$$2 \langle E_{K} \rangle = g k_{B} T \stackrel{\mathrm{3D}}{=} 3N k_{B}T \tag{1.2}$$ 
with $k_{B}$ being the Boltzmann constant, $T$ the absolute temperature and $g$ the number of translational degrees of freedom with the last equally holding for an ideal gas in a three-dimensional space. Consequently if we know the desired temperature, $T$, for the system, we can assign velocities to the particles according to the Maxwell-Boltzmann distribution
$$P(\mathbf{v}_{i})= \sqrt{\frac{m_i}{2\pi k_{B}T}} \exp{\left( {-\frac{m_i \mathbf{v}_{i}^2}{k_B T}}\right)} \tag{1.3}$$
Note here the Maxwell-Boltzmann distribution is expressed in terms of vectorial velocities ($\mathbf{v}$), unlike previously where we adressed the particle's speed ($v$) (absolute value of the velocity). Realizing that eq. 1.3 is simply a [Gaussian distribution](https://en.wikipedia.org/wiki/Normal_distribution#Standard_normal_distribution) with a mean equal to 0 and variance equal to
$$\sigma_{i}^2 = \frac{k_B T}{m_{i}} \tag{1.4}$$
The problem boils down to generating random numbers from a gaussian distribution, as the Maxwell-Boltzmann distribution can be written as
$$\mathbf{v}_{i}=\sqrt{\frac{k_B T}{m}} \mathcal{N}(0,1) \tag{1.5}$$
In order to generate random gaussian distributed numbers with mean 0 and variance 1 from randomly uniformed numbers between 0 and 1, we conduct a [Box‚ÄìMuller transformation](https://en.wikipedia.org/wiki/Box‚ÄìMuller_transform#Basic_form) of the randomly uniformed numbers $U_{1}$ and $U_{2}$ into standard normal distributed random numbers $Z_{1}$ and $Z_{2}$, through eq. 1.6 and 1.7
$$Z_{1} = \sqrt{-2 \ln U_{1}} \cos(2\pi U_{2}) \tag{1.6}$$
$$Z_{2} = \sqrt{-2 \ln U_{2}} \cos(2\pi U_{1}) \tag{1.7}$$
<br/><br/>
In the code below we have illustrated how this procedure can successfully generate initial Maxwell-Boltzmann distributed velocities for a nitrogen gas using the Molecular Overkill Engine compared to the analytical solution.

```python
data = np.loadtxt('data/velocities.txt', usecols=1, unpack=True)
data_unit = data * 1000    # nm/ps -> m/s

# Maxwell-Boltzmann distribution
m = 2*14.0067*1.660538921e-27  # Mass (atomic mass units * dalton->kg)
T = 298.15                     # Temperature (Kelvin)
kB = 1.3806503e-23             # Boltzmann constant (J/K)
def fmb(m, T, v):
    return (m/(2.*np.pi*kB*T))**1.5 * 4.*np.pi*v*v * np.exp((-m*v*v)/(2.*kB*T))

v = np.arange(0.0, 1500, 0.01)

# Initialize figure and axes
fig = plt.figure(figsize=(8.4,6))
ax1 = fig.add_subplot(111)

# Plot Data
ax1.plot(v, fmb(m, T, v), 'r--', linewidth=3)
ax1.hist(data_unit, bins=50, density=True, facecolor='green', alpha=0.70, edgecolor='purple');

# Axis modification
ax1.set_xlim(0,1500)
ax1.minorticks_on()
ax1.tick_params(axis='both',which='minor',length=5,width=2,labelsize=18)
ax1.tick_params(axis='both',which='major',length=8,width=2,labelsize=18)

# Labels and text
ax1.set_xlabel(r'Speed (m/s)', fontsize=22, labelpad=20)
ax1.set_ylabel(r'Probability', fontsize=22, labelpad=20)
ax1.set_title(r'Maxwell-Boltzmann Distribution of N$_{2}$', fontsize=22)
ax1.text(650, 0.0015, r'$f(v) = \sqrt{(\frac{m}{2 \pi k_{B} T})^{3}} 4\pi v^{2} e^{{\frac{-mv^{2}}{2k_{B}T}}}$', fontsize=20, color='b')

plt.tight_layout()
#fig.savefig('Distribution.png', hbox_inches='tight', dpi=600)
plt.show()
Markdown('''**Figure 1.1**: Histogram of the initial speed distribution of nitrogen gas (N={:.2e})
            from a molecular dynamics simulation (green) compared to the analytical 
            Maxwell-Boltzmann distribution (<span style="color:red">**--**</span>)
            with the equation written in the plot.'''.format(len(data_unit)))
```

## 2. Phase Space For The One-Dimensional Harmonic Oscillator
Newton's equations tells us if we have a set of particle positions and particle velocities at an arbitrary time, it is possible predict the positions and velocities at any one particular instant in time. Therefore, classical dynamics of a $N$-particle system can be expressed by specifying the full set of 6$N$ functions ($2N$ for each dimension). The $6N$ numbers can be regarded as a single point in a $6N$-dimensional space named the phase space. The phase space is a Cartesian space and solving Newton's equations yields $6N$ tubles at different times which is considered a *trajectory* in phase space. Consequently, classical motion can be described by the motion of a point along a trajetory in phase space! However phase space trajectories can only be visualized for a one-particle system, hence we will in this exercise be looking at a single one-dimensional harmonic oscillator in the microcanonical (*NVE*) and canonical (*NVT*) ensemble.

```python
# DATA ANALYSIS AND EXTRACTION
# NVT (Andersen)
f = open('data/harmonic_one_particle_andersen.gro', "r")
lines = f.read()
f.close()
matches = re.findall("-?[0-9]+\.[0-9]...(?![0-9])", lines)
arr = np.array(matches, dtype=float)
arr.resize(100000, 6)                         #Array of positions and velocities
vel_NVT = []
pos_NVT = []
for row in arr:
    vel_NVT.append(row[3])
    pos_NVT.append(row[0])

# NVE
f = open('data/harmonic_one_particle_no_thermostat.gro', "r")
lines = f.read()
f.close()
matches = re.findall("-?[0-9]+\.[0-9]...(?![0-9])", lines)
arr2 = np.array(matches, dtype=float)
arr2.resize(100000, 6)                         #Array of positions and velocities
vel_NVE = []
pos_NVE = []
for row in arr2:
    vel_NVE.append(row[3])
    pos_NVE.append(row[0])

# PLOTTING
# Figure Initilization
fig = plt.figure(figsize=(16,6))
scatter_axes_NVE = plt.subplot2grid((3, 8), (1, 0), rowspan=2, colspan=2)
x_hist_axes_NVE = plt.subplot2grid((3, 8), (0, 0), colspan=2, sharex=scatter_axes_NVE, frameon=False)
y_hist_axes_NVE = plt.subplot2grid((3, 8), (1, 2), rowspan=2, sharey=scatter_axes_NVE, frameon=False)

scatter_axes_NVT = plt.subplot2grid((3, 8), (1, 4), rowspan=2, colspan=2)
x_hist_axes_NVT = plt.subplot2grid((3, 8), (0, 4), colspan=2, sharex=scatter_axes_NVT, frameon=False)
y_hist_axes_NVT = plt.subplot2grid((3, 8), (1, 6), rowspan=2, sharey=scatter_axes_NVT, frameon=False)

# Axis Modifications
x_hist_axes_NVE.tick_params(axis='both', which ='both', bottom=False, labelbottom=False, left=False, labelleft=False)
y_hist_axes_NVE.tick_params(axis='both', which ='both', left=False, labelleft=False, bottom=False, labelbottom=False)
x_hist_axes_NVT.tick_params(axis='both', which ='both', bottom=False, labelbottom=False, left=False, labelleft=False)
y_hist_axes_NVT.tick_params(axis='both', which ='both', left=False, labelleft=False, bottom=False, labelbottom=False)
scatter_axes_NVT.set_xlim(-5, 5)
scatter_axes_NVT.set_ylim(-1.6, 1.6)

scatter_axes_NVE.minorticks_on()
scatter_axes_NVE.tick_params(axis='both',which='minor',length=5,width=2,labelsize=18)
scatter_axes_NVE.tick_params(axis='both',which='major',length=8,width=2,labelsize=18)
scatter_axes_NVT.minorticks_on()
scatter_axes_NVT.tick_params(axis='both',which='minor',length=5,width=2,labelsize=18)
scatter_axes_NVT.tick_params(axis='both',which='major',length=8,width=2,labelsize=18)

# Axis Labels
scatter_axes_NVE.set_ylabel(r'$\bf{v}\ \mathrm{(nm/ps)}$', fontsize=22, labelpad=20)
scatter_axes_NVE.set_xlabel(r'$\bf{x}\ \mathrm{(nm)}$', fontsize=22, labelpad=20)
scatter_axes_NVT.set_ylabel(r'$\bf{v}\ \mathrm{(nm/ps)}$', fontsize=22, labelpad=20)
scatter_axes_NVT.set_xlabel(r'$\bf{x}\ \mathrm{(nm)}$', fontsize=22, labelpad=20)
x_hist_axes_NVE.set_title(r'NVE', fontsize=30, pad=20)
x_hist_axes_NVT.set_title(r'NVT (Andersen)', fontsize=30, pad=20)

# Plotting
scatter_axes_NVE.scatter(np.array(pos_NVE)-25, vel_NVE, color='red')
x_hist_axes_NVE.hist(np.array(pos_NVE)-25, bins=34, density=True, color='green', edgecolor='purple')
y_hist_axes_NVE.hist(vel_NVE, bins=34, density=True, orientation='horizontal', color='green', edgecolor='purple')

scatter_axes_NVT.scatter(np.array(pos_NVT)-25, vel_NVT, color='red', s=0.01)
x_hist_axes_NVT.hist(np.array(pos_NVT)-25, bins=34, density=True, color='green', edgecolor='purple')
y_hist_axes_NVT.hist(vel_NVT, bins=34, density=True, orientation='horizontal', color='green', edgecolor='purple')

plt.tight_layout()
plt.subplots_adjust(wspace=0, hspace=0)

#plt.savefig('Phase_space.png', hbox_inches='tight', dpi=600)
plt.show()
Markdown('''**Figure 2.1**: Phase space for the simple classical one dimensional harmonic oscillator
            in the microcanonical (*NVE*) [*left*] and canonical (*NVT*) [*right*] ensemble using an Andersen
            thermostat with the associated positional and velocity probability distribution
            sampled using molecular dynamics.
            The force constant (spring constant) for the harmonic oscillator was 1.0 kJ/(mol*nm<sup>2</sup>).''')
```

```python
################################################
# Boltzman distribution: Potential energy part #
################################################

# Histogram of simulated position data
plt.hist(pos_NVT, bins=50, density=True, color='green', edgecolor='purple', alpha=0.70)

# Constants
m = 2*14.0067*1.660538921e-27  # Mass (atomic mass units * dalton->kg)
T = 298.15                     # Temperature (Kelvin)
kB = 1.3806503e-23             # Boltzmann constant (J/K)
constant = 1.6605391e-21       # Force constant (J/nm^2)
xe = 25                        # Equlibrium distance (nm)
beta = 1/(kB*T)                # Thermodynamic beta

# FINISH THE CODE: START
partition_function_PE = np.sqrt(np.pi*2/(constant*beta))

def probability_PE(m, T, x):
    return np.exp(-beta*0.5*constant*(x-xe)**2)/ partition_function_PE

# FINISH THE CODE: END

x = np.arange(0.0, 50, 0.001)
plt.xlim(20,30)
plt.plot(x, probability_PE(m, T, x), 'r--', linewidth=3)
plt.xlabel(r'$\bf{x}\ \mathrm{(nm)}$')
plt.ylabel(r'$\bf{P(x)}$')
plt.show()

# Self-Check: Should always equal 1!
print('The area under the curve is equal to: {:03.2f}'.format(np.trapz(probability_PE(m,T,x), x))) 
```

```python
##############################################
# Boltzman distribution: Kinetic energy part #
##############################################

# Histogram of simulated velocity data
plt.hist(vel_NVT, bins=50, density=True, color='green', edgecolor='purple', alpha=0.70)
plt.xlim(-1.5,1.5)

# Constants
m = 2*14.0067*1.660538921e-27  # Mass (atomic mass units * Dalton->kg)
T = 298.15                     # Temperature (K)
kB = 1.3806503e-23             # Boltzmann constant (J/K)
constant = 1.6605391e-21       # Force constant (J/nm^2)
xe = 25                        # Equlibrium distance (nm)
beta = 1/(kB*T)                # Thermodynamic beta

# FINISH THE CODE: START
partition_function_KE = np.sqrt(2*np.pi/(beta*m)) * 1e-3

def probability_KE(m, T, v):
    return np.exp(-beta*0.5*m*v**2*10**6)/ partition_function_KE

# FINISH THE CODE: END

v = np.arange(-2, 2, 0.001)
plt.xlim(-1.5,1.5)
plt.plot(v, probability_KE(m, T, v), 'r--', linewidth=3)
plt.xlabel(r'$\bf{v}\ \mathrm{(nm/ps)}$')
plt.ylabel(r'$\bf{P(v)}$')
plt.show()

# Self-Check: Should always equal 1!
print('The area under the curve is equal to: {:3.2f}'.format(np.trapz(probability_KE(m,T,v), v)))
```
